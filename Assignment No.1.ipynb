```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
r1 = '/Users/no/Documents/r1.wav'
r1sig , r1sr = librosa.load (r1,mono=True,sr=None)
r1time=np.arange (0,r1sig.shape[0])/r1sr
```

```python
plt.plot (r1time,r1sig)
plt.xlabel ('Time (s)')
plt.ylabel ('Amplitude')
plt.title ('R1 Waveform')
```

```python
plt.plot (r1time,r1sig)
plt.xlabel ('Time (s)')
plt.ylabel ('Amplitude')
plt.title ('R1 Waveform')
plt.xlim (15,25)
plt.ylim (-0.5,0.5)
```

```python
r1spec = librosa.amplitude_to_db (np.abs(librosa.stft (r1sig)), ref=np.max)
plt.figure (figsize= (10, 4))
librosa.display.specshow (r1spec, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title ('R1 Linear-frequency power spectrogram')
```

```python
plt.figure (figsize= (10, 6))
librosa.display.specshow (r1spec, y_axis='log')
plt.colorbar (format='%+2.0f dB')
plt.title ('R1 Log-frequency power spectrogram')
```
