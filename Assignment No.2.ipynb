```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
file = '/Users/no/Documents/avm.wav'
sig , sr = librosa.load (file,mono=True,sr=None)
```

```python
plt.figure (figsize=(16, 10))
plt.title ('Log-frequency power spectrogram')
plt.subplot (2,2,1)
r1Spec = librosa.amplitude_to_db (np.abs(librosa.stft (sig,n_fft=2048, hop_length=32,win_length=1024)), ref=np.max)
librosa.display.specshow(r1Spec, y_axis='log')
plt.colorbar (format='%+2.0f dB')
```

```python
plt.figure (figsize=(16, 10))
plt.subplot (2,2,2)
S = librosa.feature.melspectrogram (y=sig, sr=sr,n_fft=2048,n_mels=128)
log_S = librosa.power_to_db (S, ref=np.max)
librosa.display.specshow (log_S, sr=sr, x_axis='time', y_axis='mel')
```

```python
plt.figure (figsize= (16, 10))
plt.subplot (2,2,3)
sig_harmonic, sig_percussive = librosa.effects.hpss(sig)
chromagram = librosa.feature.chroma_cqt (y=sig_harmonic,sr=sr)
librosa.display.specshow (chromagram, sr=sr, x_axis='time', y_axis='chroma', vmin=0, vmax=1)
plt.title ('Chromagram')
plt.colorbar ( )
```

```python
plt.figure (figsize= (16, 10))
plt.subplot (2,2,4)
mfcc = librosa.feature.mfcc (S=log_S, n_mfcc=13)
librosa.display.specshow (mfcc [1:12,:])
plt.ylabel ('MFCC')
plt.colorbar ( )
```
